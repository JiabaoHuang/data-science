## 透视表功能

pd.pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, dropna=True, margins_name='All') <p>
data:指定需要构造透视表的数据集 <p>
values:指定需要拉入“数值”框的字段列表<p>
index:指定需要拉入“行标签”框的字段列表 <p>
columns:指定需要拉入“列标签”框的字段列表 <p>
aggfunc:指定数值的统计函数，默认为统计均值，也可以指定numpy模块中的其他统计函数 <p>
fill_value:指定一个标量，用于填充缺失值 <p>
margins:bool类型参数，是否需要显示行或列的总计值，默认为False <p>
dropna:bool类型参数，是否需要删除整列为缺失的字段，默认为True <p>
margins_name:指定行或列的总计名称，默认为All<p>

## 分组汇总
groupby“方法”:用于汇总前，设定被分组的变量 <p>
    aggregate“方法”:可基于groupby的结果做进一步的统计汇总<p>
说明:在aggregate阶段，需以字典的形式传递参数，用于选择被统计的变量和对应的统计方法

```python
import numpy as np
```

```python
grouped = data6.groupby(by = ['color','cut']) #通过groupby指定分组变量
result = grouped.aggregate({'color':np.size, 'carat':np.min,'price':np.mean, 'table':np.max}) #通过aggregate进行统计汇总
result.head()
```

```python
# 通过DataFrame调整变量的顺序
result = pd.DataFrame(result, columns=['price', 'color', 'table', 'carat'])
result
```

```python
# 数据集的重命名
result.rename(columns={'color': 'count', 'price':'ave_price', 'carat':'min_carat', 'table':'max_surface_width' })
```

## 列连接

pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'))


<p>
left:指定需要连接的主 <p>
right:指定需要连接的辅表<p>
how:指定连接方式，默认为inner内连，还有其他选项，如左连left、右连right和外连outer <p>
on:指定连接两张表的共同字段<p>
left_on:指定主表中需要连接的共同字段<p>
right_on:指定辅表中需要连接的共同字段 <p>
left_index:bool类型参数，是否将主表中的行索引用作表连接的共同字段，默认为False <p>
right_index:bool类型参数，是否将辅表中的行索引用作表连接的共同字段，默认为False <p>
sort:bool类型参数，是否对连接后的数据按照共同字段排序，默认为False <p>
suffixes:如果数据连接的结果中存在重叠的变量名，则使用各自的前缀进行区分<p>

```python
# 构造数据集
df3 = pd.DataFrame({'id':[1,2,3,4,5],'name':['张三','李四','王二','丁一','赵五'],'age':[27,24,25,23,25],'gender':['男','男','男','女','女']}) 
df4 = pd.DataFrame({'Id':[1,2,2,4,4,4,5], 'score':[83,81,87,75,86,74,88],'kemu':['科目1','科目1','科目2','科目1','科目2','科目3','科目1']}) 
df5 = pd.DataFrame({'id':[1,3,5],'name':['张三','王二','赵五'],'income':[13500,18000,15000]})

print(df3)
print(df4)
print(df5)
```

```python
# 首先合并df3和df4
merge1 = pd.merge(df3,df4, how='left', left_on= 'id', right_on= 'Id')
merge1
```

```python
# 再继续合并df5
merge2 = pd.merge(merge1,df5 ,how='left')
merge2
```
