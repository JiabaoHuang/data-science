```python
import bs4
import re
import requests
import pandas as pd
```

```python
url = r'https://sh.lianjia.com/ershoufang/pudong/pgl/'

headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36'}
```

```python
response = requests.get(url, headers = headers)
response #与课程不同，视频中链家具有反爬功能，无法建立链接（403），自己操作可以建立连接
```

```python
soup = bs4.BeautifulSoup(response.text)
```

```python
region = [ i.text.strip() for i in soup.findAll(name = 'a', attrs={'data-el':'region'})]
region
```

```python
houseinfo = [ i.text.split('|') for i in soup.findAll(name='div', attrs={'class':'houseInfo'})]
houseinfo
```

```python
Type = [i[0].strip() for i in houseinfo]
area = [float(i[1].strip()[:-2]) for i in houseinfo ]
direction = [ i[2].strip() for i in houseinfo ]
decoration = [ i[3].strip() for i in houseinfo ]
floor = [ i[4].strip() for i in houseinfo ]
construction_time = [ i[5].strip()[:-2] for i in houseinfo ]
architecture = [ i[-1] for i in houseinfo ] #观察数据有一行缺少建筑年份
```

```python
totalprice = [ int(i.text[:-1]) for i in soup.findAll(name='div', attrs={'class':'totalPrice'})]
unitprice = [ int(i.text[2:-4]) for i in soup.findAll(name='div', attrs={'class':'unitPrice'}) ]
```

```python
pd.DataFrame({'名称':region ,'户型':Type ,'面积':area ,'朝向':direction ,'装修':decoration ,'楼层':floor ,'建筑年份':construction_time ,'总价/元':totalprice ,'单价/元': unitprice})
```
