```python
import re
import  requests

url = r'http://www.redbull.com.cn/about/branch'

response = requests.get(url)
requests
```

```python
response.text
```

```python
# 提取公司名称
compony = re.findall('<h2>(.*?)</h2>',response.text)
```

```python
# 提取公司地址
address = re.findall("<p class=\'mapIco\'>(.*?)</p>",response.text)
```

```python
import bs4
soup = bs4.BeautifulSoup(response.text)
```

```python
mail = [i.text for i in soup.findAll(name='p',attrs={'class':'mailIco'})]
tel = [ i.text for i in soup.findAll(name='p',attrs={'class':'telIco'})]
```

```python
import pandas as pd
pd.DataFrame({'compony':compony,'address':address,'mail':mail,'tel':tel})
```
