```python
url = r'http://tianqi.2345.com/t/wea_history/js/201910/58362_201910.js'

import requests
import pandas

response = requests.get(url)
response
```

```python
response.text
```

```python
import re
date = re.findall("ymd:'([\d-]*?)'",response.text)
highdegree = re.findall("bWendu:'([\d-]+)℃'",response.text)
lowdegree = re.findall("yWendu:'([\d-]+)℃'",response.text)
weather = re.findall("tianqi:'(.*?)'",response.text)

pandas.DataFrame({'date':date, 'highdegree/℃':highdegree, 'lowdegree/℃':lowdegree, 'weather':weather})
```
